```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split

cancer=load_breast_cancer()

print(cancer.keys())
```

```python
X_train, X_test, y_train, y_test=train_test_split(cancer.data, cancer.target, random_state=0)
```

```python
grb=GradientBoostingClassifier(random_state=0)
grb.fit(X_train, y_train)
```

```python
print(f"Accuracy on training set: {grb.score(X_train, y_train):.3f}")
print(f"Accuracy on test: {grb.score(X_test, y_test):.3f}")
```

```python
grb=GradientBoostingClassifier(random_state=0, max_depth=1)
grb.fit(X_train, y_train)

print(f"Accuracy on training set: {grb.score(X_train, y_train):.3f}")
print(f"Accuracy on test: {grb.score(X_test, y_test):.3f}")
```

```python
grb=GradientBoostingClassifier(random_state=0, learning_rate=0.01)
grb.fit(X_train, y_train)

print(f"Accuracy on training set: {grb.score(X_train, y_train):.3f}")
print(f"Accuracy on test: {grb.score(X_test, y_test):.3f}")
```

```python
train_acc=[]
test_acc=[]

model_depth=range(1,10)
for depth in model_depth:
    grb=GradientBoostingClassifier(random_state=0, max_depth=depth)
    grb.fit(X_train, y_train)
    train_acc.append(grb.score(X_train, y_train))
    test_acc.append(grb.score(X_test, y_test))
plt.plot(model_depth, train_acc, label="Training Accuracy")
plt.plot(model_depth, test_acc, label="Test Accuracy")
plt.xlabel("model_depth")
plt.ylabel("Accuracy")
plt.legend()
```

```python
train_acc=[]
test_acc=[]
learn=[]
learn_rate=range(1,10)
for depth in learn_rate:
    grb=GradientBoostingClassifier(random_state=0, learning_rate=depth/1000)
    grb.fit(X_train, y_train)
    train_acc.append(grb.score(X_train, y_train))
    test_acc.append(grb.score(X_test, y_test))
    learn.append(depth/1000)
plt.plot(learn, train_acc, label="Training Accuracy")
plt.plot(learn, test_acc, label="Test Accuracy")
plt.xlabel("model_depth")
plt.ylabel("Accuracy")
plt.legend()
```

```python
grb=GradientBoostingClassifier(random_state=0, learning_rate=np.exp(1)/1000)
grb.fit(X_train, y_train)

print(f"Accuracy on training set: {grb.score(X_train, y_train):.3f}")
print(f"Accuracy on test: {grb.score(X_test, y_test):.3f}")
```
