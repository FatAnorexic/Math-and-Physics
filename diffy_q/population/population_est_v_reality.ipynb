# Preamble
For this model, the goal was to look at a simple differential model measuring us population growth. Where $\frac{dP}{dt}=kP$ and $P(0)=3.9$

## Solution

To solve this, we use the equation from earlier $P(t)=P_0\ e^{kt}$ and solve for $k$ using $P(10)=5.3$
$$5.3=3.9\ e^{k10}$$
$$\frac{5.3}{3.9}=e^{k10}$$
$$\ln(\frac{5.3}{3.9})=k10$$
$$k\approx 0.03067$$

```python
#List of actual US poplation 1790-2000| source: www.cencus.gov

actual_population=[
    3.9, 5.3, 7.2, 9.6, 13, 17, 23, 31, 39, 50, 63, 76, 91, 106,
    123, 132, 151, 179, 203, 227, 249, 281
    ]

#Start year
t_0=1790

#List to hold predicted values from our model
pred_population=[]
t=[x*10+t_0 for x in range(len(actual_population))]
```

```python
def population_model(t):
    return (3.9*np.exp(0.03067*t))
```

```python
for time in range(len(actual_population)):
    pred_population.append(population_model(time*10).item())
    
```

```python
plt.plot(t, actual_population, linestyle='dotted', label="Actual Population")
plt.plot(t, pred_population, label="Predicted Population")
plt.ylabel("Population")
plt.xlabel("Year")
plt.legend()
```

# Results
As you can see, the model we derived earlier works well up until about 1850-1870, where it deviates drastically. This is in part due to the fact that $k$ can also
vary. In later models we will explore ways to account for variable rate of growth, but for now, this is a solid example of a very basic differential model.
