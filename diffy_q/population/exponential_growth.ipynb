# Derivation
Assuming $r\gt0$ and that the rate of growth is proportional to the value of $y$ then $$\frac{dy}{dt}=ry$$

Solving this: $$\frac{dy}{y}=rdt\Rightarrow\ln(y)=rt+C\Rightarrow y(t)=Ce^{rt}$$

letting $y(0)=y_0$ then $C=y_0$ and the solution is 

$$Expontential\;Growth\\
y(t)=y_0e^{rt}$$

in the example below, we set $r=0.3$

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

st=np.arange(0,10,0.1)
sy=np.arange(0,250,10)

ts, ys=np.meshgrid(st, sy)

dy=0.3*ys
dt=np.ones(dy.shape)
def norm(dt, dy):
    return np.sqrt(dt**2+dy**2)

n=norm(dt, dy)

def expon_growth(r, t, init):
    return init*np.exp(r*t)

t=np.linspace(0,10,100)
plt.quiver(ts, ys, dt/n, dy/n, headaxislength=0)
plt.plot(t, expon_growth(0.3, t, 10))
```
