### 1. $y'+3y=t+e^{-2t}$

```python
import numpy as np
import matplotlib.pyplot as plt

st=np.arange(0,10,0.3)
sy=np.arange(0,10,0.3)

ts, ys=np.meshgrid(st, sy)

dy=ts+np.exp(-2*ts)-3*ys
dt=np.ones(dy.shape)
def norm(dt, dy):
    return np.sqrt(dt**2+dy**2)
n=norm(dt, dy)

plt.quiver(ts, ys, dt/n, dy/n, headaxislength=0)
```

```python
def one_sol(t,c):
    return c*np.exp(-3*t)+t/3-1/9+np.exp(-2*t)

time=np.linspace(0,10, 100)
c=[-1,0,1]
for con in c:
    plt.plot(time, one_sol(time, con), label=rf'$c={con}$')
plt.legend()

import pandas as pd

data=pd.DataFrame({'y(t)':[one_sol(t, 1) for t in time], 't':[t for t in time]})
print(data)
```

### (4) $y'+\frac{1}{t}y=3\cos{2t}\quad t\gt0$

```python
st=np.arange(0.1, 10, 0.3)
sy=np.arange(-3, 10, 0.3)
ts, ys=np.meshgrid(st, sy)

dy=3*np.cos(2*ts)-ys/ts
dt=np.ones(dy.shape)
n=norm(dy, dt)

plt.quiver(ts, ys, dt/n, dy/n, headaxislength=0)

time=np.linspace(0.1, 10, 100)
def four_sol(t, c):
    return 3/2*np.sin(2*t)+np.cos(2*t)/(4*t)+c/t

const=[-0.5, 0, 0.5]
for c in const:
    plt.plot(time, four_sol(time, c), label=rf'c={c}')
plt.legend()

data=pd.DataFrame({'t':[t for t in time],'c=-0.5':[round(four_sol(t, -0.5),3) for t in time], 'c=0':[round(four_sol(t, 0),3) for t in time], 'c=0.5':[round(four_sol(t, 0.5),3) for t in time]})
print(data)
```
