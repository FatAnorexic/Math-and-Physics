### 1. $y'+3y=t+e^{-2t}$

```python
import numpy as np
import matplotlib.pyplot as plt

st=np.arange(0,10,0.3)
sy=np.arange(0,10,0.3)

ts, ys=np.meshgrid(st, sy)

dy=ts+np.exp(-2*ts)-3*ys
dt=np.ones(dy.shape)
def norm(dt, dy):
    return np.sqrt(dt**2+dy**2)
n=norm(dt, dy)

plt.quiver(ts, ys, dt/n, dy/n, headaxislength=0)
```

```python
def one_sol(t,c):
    return c*np.exp(-3*t)+t/3-1/9+np.exp(-2*t)

time=np.linspace(0,10, 100)
c=[-1,0,1]
for con in c:
    plt.plot(time, one_sol(time, con), label=rf'$c={con}$')
plt.legend()
```
